## Model and bins

```python
import pickle

files = [
    "../models/logistic_model.pkl",
    "../models/woe_bins.pkl",
]

for file in files:
    print("=" * 60)
    print(file)

    with open(file, "rb") as f:
        obj = pickle.load(f)

    print(type(obj))

    if hasattr(obj, "keys"):
        print("Dictionary keys:")
        print(list(obj.keys())[:10])

    else:
        print(obj)
```
