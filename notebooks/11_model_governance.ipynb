# 11. Model Governance Discussion

## Objective

This notebook documents how the credit risk scorecard model would be governed in a real financial institution.

It does not perform additional modelling.  
The purpose is to demonstrate awareness of:

- model ownership;
- model risk management;
- model documentation;
- independent validation;
- regular monitoring;
- regulatory and compliance expectations;
- implementation controls;
- limitations and future improvements.



## 1. Model Purpose and Intended Use

The model developed in this project is a credit risk classification / scorecard model.

The model estimates the likelihood that a loan will result in a bad outcome, defined as:

```text
Charged Off = Bad
Fully Paid = Good
```

The model is intended for educational and case-study purposes and should be interpreted as a historical loan outcome model rather than a regulatory 12-month PD model.

### Intended use

The model could support:

- borrower risk ranking;
- credit application screening;
- scorecard-based segmentation;
- cutoff strategy analysis;
- challenger model benchmarking;
- portfolio risk monitoring.

### Not intended use

- an approved regulatory IRB PD model;
- an IFRS 9 lifetime expected credit loss model;
- a production credit decision engine;
- a final automated approval / rejection model without additional governance, validation and compliance review.


## 2. Model Ownership and Responsibilities

In a real institution, clear model ownership is required.

### Model owner

The model owner is responsible for:

- approving the business purpose of the model;
- defining intended use and limitations;
- ensuring the model remains fit for purpose;
- approving model changes;
- reviewing monitoring results;
- escalating material performance deterioration.

For a credit risk scorecard, the model owner would typically sit within Credit Risk, Risk Analytics or Credit Risk Modelling.

### Model developer

The model developer is responsible for:

- data preparation and target definition;
- feature engineering;
- WOE / IV analysis;
- model development;
- performance testing;
- documentation of methodology and assumptions;
- reproducible code and artifacts;
- handover to validation and implementation teams.

### Model validation function

An independent validation team should review:

- conceptual soundness;
- data quality;
- target definition;
- variable selection;
- model methodology;
- performance and stability;
- calibration;
- limitations;
- implementation readiness;
- monitoring framework.

### Business users

Business users are responsible for:

- understanding how the scorecard should be used;
- applying cutoffs according to approved risk appetite;
- avoiding use outside approved scope;
- escalating unusual model behaviour.


## 3. Model Documentation Requirements

A production-grade model would require formal model documentation.

The documentation should include the following sections.

### 3.1 Executive Summary

- model objective;
- intended use;
- target population;
- model type;
- key performance results;
- key limitations;
- recommendation for use.

### 3.2 Data Description

- data source;
- data extraction date;
- observation period;
- performance period;
- inclusion and exclusion criteria;
- target definition;
- data quality checks;
- missing values;
- outlier treatment;
- leakage assessment.

### 3.3 Target Definition

The case-study target is based on final loan status:

```text
Bad = Charged Off
Good = Fully Paid
```

This should be clearly documented as a historical loan outcome target, not a regulatory 12-month default flag.

### 3.4 Modelling Methodology

Documentation should cover:

- time-based train / validation / OOT split;
- maturity bias treatment;
- feature engineering;
- WOE transformation;
- IV analysis;
- correlation reduction;
- business review exclusions;
- logistic regression methodology;
- score scaling;
- cutoff scenarios;
- challenger model comparison.

### 3.5 Model Results

The documentation should include:

- AUROC;
- Gini;
- KS;
- Brier score;
- decile analysis;
- calibration;
- PSI;
- score distribution;
- cutoff analysis.

### 3.6 Limitations

Limitations should be explicitly documented:

- public Kaggle dataset;
- no reject population;
- no formal application / bureau production data lineage;
- no regulatory 12-month default definition;
- no clear observation and performance window;
- limited macroeconomic information;
- no profitability / LGD / EAD information;
- no fairness or protected-class analysis;
- no production implementation testing.

### 3.7 Monitoring Plan

The documentation should define what will be monitored, how often, and what actions are triggered by deterioration.


## 4. Model Risk Assessment

A formal model risk assessment should be performed before production use.

The assessment should consider:

### Materiality

- Does the model influence credit approval?
- Does it affect pricing?
- Does it affect provisioning or capital?
- What portfolio size is affected?
- What is the financial impact of wrong decisions?

### Complexity

- Logistic regression scorecard: relatively low to medium complexity;
- WOE transformation: transparent and standard in credit risk;
- XGBoost challenger: higher complexity and lower transparency.

### Usage Risk

Risk is higher if the model is used for:

- automated rejection;
- pricing;
- regulatory capital;
- impairment provisioning;
- high-volume origination.

Risk is lower if the model is used for:

- ranking;
- monitoring;
- analyst support;
- challenger benchmarking.

### Model Risk Rating

For this case study, the baseline WOE Logistic Regression would likely be considered lower risk than the XGBoost challenger due to stronger interpretability and easier governance.


## 5. Independent Validation Expectations

Independent validation should be performed before the model is approved for use.

The validation should review:

### Conceptual Soundness

- Is the model objective clearly defined?
- Is the target variable appropriate?
- Is the modelling approach suitable?
- Are exclusions and assumptions justified?
- Are WOE transformations logical and stable?

### Data Validation

- Are source data fields reliable?
- Are application-time variables separated from post-outcome variables?
- Is leakage appropriately controlled?
- Are missing values and outliers handled consistently?
- Are latest vintages sufficiently seasoned?

### Performance Validation

- Train / validation / OOT AUROC;
- Gini and KS;
- score decile separation;
- calibration;
- stability across vintages;
- challenger model comparison.

### Stability Validation

- score PSI;
- characteristic PSI;
- population drift;
- performance drift;
- calibration drift.

### Implementation Validation

- are model coefficients implemented correctly?
- are WOE mappings implemented correctly?
- is score scaling correct?
- are cutoffs aligned with approved policy?
- are all production transformations consistent with development code?

### Validation Outcome

Independent validation should result in one of the following:

- approved;
- approved with conditions;
- remediation required;
- rejected for use.


## 6. Regular Monitoring Framework

A model should not be considered complete after development.  
Regular monitoring is required to confirm that it remains fit for purpose.

### Monitoring Frequency

A practical monitoring schedule could be:

| Monitoring Area | Frequency |
|---|---|
| Data quality checks | Monthly |
| Score distribution | Monthly |
| Approval rate by cutoff | Monthly |
| Characteristic PSI | Monthly / Quarterly |
| Score PSI | Monthly / Quarterly |
| Bad rate by score band | Quarterly |
| AUROC / Gini / KS | Quarterly or when outcomes mature |
| Calibration | Quarterly or semi-annually |
| Full model review | Annually |

### Key Monitoring Metrics

The monitoring dashboard should include:

- application volume;
- approval rate;
- average score;
- average predicted PD;
- observed bad rate;
- score distribution;
- score PSI;
- characteristic PSI;
- decile bad rates;
- AUROC / Gini / KS;
- calibration by score band;
- overrides and manual decisions;
- missing values and data quality issues.

### Example Thresholds

Illustrative monitoring thresholds:

| Metric | Green | Amber | Red |
|---|---:|---:|---:|
| Score PSI | < 0.10 | 0.10 - 0.25 | > 0.25 |
| Characteristic PSI | < 0.10 | 0.10 - 0.25 | > 0.25 |
| AUROC drop vs development | < 0.03 | 0.03 - 0.05 | > 0.05 |
| Calibration gap | < 3 p.p. | 3 - 5 p.p. | > 5 p.p. |

Thresholds should be adapted to the institution's model risk policy and portfolio characteristics.

### Escalation Actions

If monitoring breaches occur, possible actions include:

- investigate data quality;
- review portfolio mix changes;
- recalibrate PDs;
- adjust cutoffs;
- rebuild WOE bins;
- redevelop the model;
- temporarily restrict model use;
- escalate to model risk committee.


## 7. Regulatory and Compliance Considerations

A real credit risk model must comply with relevant regulatory, legal and internal policy requirements.

Depending on jurisdiction and use case, this may include:

- banking supervision expectations for model risk management;
- internal model governance policy;
- credit risk management policy;
- data protection requirements;
- fair lending / anti-discrimination expectations;
- explainability requirements for automated decisions;
- auditability and reproducibility standards;
- change management and approval workflows.

### Key regulatory themes

Important governance themes include:

- clear model purpose;
- documented assumptions;
- traceable data lineage;
- avoidance of data leakage;
- independent validation;
- regular monitoring;
- human oversight;
- appropriate use of challenger models;
- explainability of credit decisions;
- audit trail for model changes.


## 8. Implementation Controls

If the model were implemented in production, several controls would be required.

### Technical Controls

- version-controlled code;
- controlled model artifacts;
- approved WOE binning mappings;
- locked model coefficients;
- reproducible scoring pipeline;
- automated data quality checks;
- logging of input data and scores;
- monitoring of missing and out-of-range values.

### Business Controls

- approved score cutoffs;
- documented override rules;
- manual review process;
- periodic review of approval rates;
- approval from credit committee or model governance committee;
- clear ownership of cutoff changes.

### Change Management

Any material model change should follow a controlled process:

1. change request;
2. impact assessment;
3. redevelopment or recalibration;
4. independent validation;
5. approval;
6. production deployment;
7. post-implementation monitoring.

Examples of material changes:

- new variables;
- new coefficients;
- new WOE bins;
- new target definition;
- new population;
- new data source;
- new score cutoffs with material business impact.


## 9. Challenger Model Governance

Notebook 08 included an XGBoost / gradient boosting challenger.

Although the challenger showed some performance improvement, it is not automatically selected as the primary model.

### Why keep logistic regression as primary?

The WOE Logistic Regression scorecard is preferred because it is:

- more transparent;
- easier to explain;
- easier to validate;
- easier to monitor;
- easier to convert into score points;
- more aligned with traditional credit risk scorecard governance.

### Challenger role

The challenger model can be retained for:

- benchmarking;
- monitoring performance gap;
- identifying nonlinear relationships;
- future redevelopment;
- identifying interactions or feature engineering opportunities.

### Additional governance required for XGBoost

If the challenger were considered for production, additional work would be needed:

- hyperparameter tuning;
- overfitting assessment;
- explainability analysis;
- SHAP or feature attribution;
- monotonicity and reason-code design;
- implementation testing;
- stronger validation and monitoring framework.


## 10. Model Limitations

The main model limitations are:

1. **Dataset limitation**  
   The dataset is public and preprocessed. It is not a controlled internal bank data source.

2. **Target limitation**  
   The target is historical final outcome, not a formally defined 12-month regulatory default.

3. **Observation window limitation**  
   The dataset does not provide a formal observation and performance window.

4. **Reject inference limitation**  
   The dataset contains accepted and originated loans only. Rejected applicants are not observed.

5. **Profitability limitation**  
   The model does not include LGD, EAD, pricing, fees, funding cost or profitability.

6. **Fairness limitation**  
   No formal fair-lending or disparate-impact analysis was performed.

7. **Macroeconomic limitation**  
   The model does not include macroeconomic variables or stress scenario sensitivity.

The selected WOE Logistic Regression model is preferred as the baseline due to its interpretability, stability and suitability for scorecard governance.
